```python
import sympy
import sympy.plotting as sym_plot
from sympy import I

sympy.init_printing()
a = 0.6
b = 5.5
c = 10
d = 5
x = sympy.Symbol('x')
f_x = a*(x**3) + b*(x**2) + c*(x**1) - d
#f_x = x**2 - 4
%matplotlib inline
g1 = sym_plot.plot(f_x,  (x, -7, 2), title=" ", line_color = "red", legend=True, xlabel="X", ylabel="Y")

df_x = f_x.diff(x)
g2 = sym_plot.plot(f_x, df_x, (x, -7, 0), line_color = "blue", legend=True)

koren_df_x = sympy.solve(df_x, x)
print(" : " + str(koren_df_x))

if koren_df_x[0] - 1 < 0:
    print("   :     " + str(koren_df_x[0]) + "   " + str(koren_df_x[1]) + "  ")
    print("     " + str(koren_df_x[0]) + "  " + str(koren_df_x[1]))
else:
    print("   :     " + str(koren_df_x[0]) + "   " + str(koren_df_x[1]) + "  ")
    print("     " + str(koren_df_x[0]) + "  " + str(koren_df_x[1]))

koren_f_x = sympy.solve(f_x, x)

result = []
for item in koren_f_x:
    ff = item.as_real_imag()
    result.append(float(ff[0]))
print(": " + str(result))

if len(result) == 3:
    if result[0] - 1 < 0:
        print("        " + str(result[0]) + "   " + str(result[1]) + "  " + str(result[2]))
        print("     " + str(result[0]) + "  " + str(result[1]) + "   " + str(result[2]) + "   ")
    else:
        print("        " + str(result[0]) + "   " + str(result[1]) + "  " + str(result[2]))
        print("     " + str(result[0]) + "  " + str(result[1]) + "   " + str(result[2]) + "   ")
if len(result) == 1:
    if result[0] - 1 < 0:
        print("        " + str(result[0]))
        print("     " + str(result[0]) + "   ")
    else:
        print("        " + str(result[0]))
        print("     " + str(result[0]) + "   ")
```
